# 2. Line Plot(선 그래프)
* 시간에 따른 변화의 추이를 시각화하는데 유용

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 2-1. 그래프 그리기

```python
# (1,3),(8,10) 두 점을 연결하는 선 그래프 그리기
x = [1,8]
y = [3,10]
plt.plot(x,y)
```

```python
# 두 점만 그리기 (marker : o (rings))
plt.plot(x,y,'o')
plt.show()
```

```python
x=np.arange(1,11)
y=x*2
plt.plot(x,y,'o')
plt.plot(x,y)
plt.show()
```

```python
# 점과 선 둘 다 그리기
plt.plot(x,y,marker='o')
plt.show()
```

## 2-2. Marker,Line, Color
* Marker : https://matplotlib.org/stable/api/markers_api.html?highlight=marker
* Line : https://matplotlib.org/3.0.3/gallery/lines_bars_and_markers/line_styles_reference.html
* Color : https://matplotlib.org/stable/gallery/color/named_colors.html


```python
# Marker
plt.plot(x,y,marker='*',ms=10)
plt.show()
```

```python
# Line Style
plt.plot(x,y,ls=':')
plt.plot(y,x,ls='-.')
plt.show()
```

```python
# Color
plt.plot(y,x,color='r')
plt.plot(x,y,c='b',linewidth=100)
plt.show()
```

```python
# Format String
plt.plot(x,y,'o:r')
plt.show()
```

```python
#그래프 분리하기
plt.plot(x,y,'o:r')
plt.figure()
plt.plot(x,y,'*--b')
plt.show()
```

```python
# 그래프 사이즈 변경하기
plt.figure(figsize=(5,5))
plt.plot(x,y)
plt.figure(figsize=(10,5))
plt.plot(y,x)
plt.show()
# 단위가 인치
```

## 2-3. Title, Label, Tick, Legend

```python
# Title 설정
plt.title('MooMinKIM')
plt.plot(x,y)
plt.show()
```

```python
plt.title('MooMinKIM',size=1000)
plt.plot(x,y)
plt.show()
```

```python
# Label 설정
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(x,y)
plt.show()
```

```python
# Tick 설정
plt.xticks(rotation=90)
plt.yticks(rotation=90)
plt.plot(x,y)
plt.show()
```

```python
# legend(범례) 설정

plt.plot(x,y)
plt.plot(y,x)
plt.legend(['(x,y)','(y,x)'])
plt.show()
```

```python
# legend(범례) 설정
plt.plot(x,y,label='(x,y)')
plt.plot(y,x,label='(y,x)')
plt.legend()
plt.show()    
```

## 2-4. Grid

```python
# Grid
plt.plot(x,y)
plt.grid()
plt.show()
```

```python
# Grid
plt.plot(x,y)
plt.grid(axis='x')
plt.show()
```

```python
# Grid 속성 변경하기
plt.plot(x,y)
plt.grid(ls=':',lw=2,c='r')
plt.show()
```
